```python
import tparbressource as tp 
from sklearn import tree
import seaborn as sb
import matplotlib.pyplot as plt
```

2-

```python
from sklearn.model_selection import StratifiedShuffleSplit
res = tp.rand_checkers(n1=23*8, n2=23*8, n3=23*8, n4=23*8, sigma=0.1)
print(res.shape)
print(res[:5,:])
X = res[:,:2]
y = res[:,-1]
sss = StratifiedShuffleSplit(n_splits=2, train_size=456 ,test_size = 736-456)
sss.get_n_splits(X, y)
for train_index, test_index in sss.split(X, y):
    X_train, X_test = X[train_index], X[test_index]
    y_train, y_test = y[train_index], y[test_index]
print(X_train.shape)
print(X_test.shape)

ax = sb.countplot(y = y_test)
plt.show()
```

```python
from sklearn.metrics import accuracy_score

maxdepth = 15
# GINI
err = []
for d in range(1,maxdepth+1):
    clf = tree.DecisionTreeClassifier(criterion="gini",max_depth=d)
    clf = clf.fit(X_train,y_train)
    y_pred = clf.predict(X_test)
    err.append(accuracy_score(y_test,y_pred)) 
plt.scatter(range(1,len(err)+1),err)    
plt.show()

#Entropy
err = []
for d in range(1,maxdepth+1):
    clf = tree.DecisionTreeClassifier(criterion="entropy",max_depth=d)
    clf = clf.fit(X_train,y_train)
    y_pred = clf.predict(X_test)
    err.append(accuracy_score(y_test,y_pred)) 
plt.scatter(range(1,len(err)+1),err)    
plt.show()
    
```

```python
tp.plot_2d(X_train,y_train)
plt.show()
tp.plot_2d(X_test,y_test)
plt.show()
```
