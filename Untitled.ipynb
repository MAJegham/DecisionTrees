```python
import tparbressource as tp  #seed is fixed inside the .py file 

import graphviz 
from sklearn import tree
import seaborn as sb
import matplotlib.pyplot as plt
from sklearn.model_selection import StratifiedShuffleSplit
from sklearn.metrics import accuracy_score
import os
import datetime
from sklearn import datasets

```

2-

```python
n=23
res = tp.rand_checkers(n1=n*8, n2=n*8, n3=n*8, n4=n*8, sigma=0.1,random_state=46)
#print(res.shape)
#print(res[:5,:])
X = res[:,:2]
y = res[:,-1]
sss = StratifiedShuffleSplit(n_splits=5, train_size=456 ,test_size = 736-456, random_state=46)
sss.get_n_splits(X, y)
for train_index, test_index in sss.split(X, y):
    X_train, X_test = X[train_index], X[test_index]
    y_train, y_test = y[train_index], y[test_index]
print(X_train.shape)
print(X_test.shape)

plt.figure(figsize = (12,5))

plt.subplot(121)
plt.title("Distribution of classes in train set")
plt.ylabel("classes")
sb.countplot(y = y_train)

plt.subplot(133)
plt.title("Distribution of classes in test set")
plt.ylabel("classes")
sb.countplot(y = y_test)

plt.show()
```

```python
def gini_entropy_plot(X_train,y_train,X_test,y_test,maxdepth = 2):
    plt.figure(figsize = (12,5))

    # GINI
    err = []
    for d in range(1,maxdepth+1):
        clf = tree.DecisionTreeClassifier(criterion="gini",max_depth=d,random_state=25)
        clf = clf.fit(X_train,y_train)
        y_pred = clf.predict(X_test)
        err.append(1-accuracy_score(y_test,y_pred)) 
    g_depth = err.index(min(err)) + 1
    g_err = min(err)
    plt.subplot(121)
    plt.title("percentage of error using gini_index\n with respect to tree's depth")
    plt.xlabel("tree depth")
    plt.ylabel("error percentage")
    plt.scatter(range(1,len(err)+1),err)

    #Entropy
    err = []
    for d in range(1,maxdepth+1):
        clf = tree.DecisionTreeClassifier(criterion="entropy",max_depth=d,random_state=25)
        clf = clf.fit(X_train,y_train)
        y_pred = clf.predict(X_test)
        err.append(1-accuracy_score(y_test,y_pred)) 
    e_depth = err.index(min(err)) + 1
    e_err = min(err)
    plt.subplot(122)
    plt.title("percentage of error using entropy\n with respect to tree's depth")
    plt.xlabel("tree depth")
    plt.ylabel("error rate")
    plt.scatter(range(1,len(err)+1),err)    

    plt.show()

    print('GINI index criterion:')
    print('Best achieved error : %.4f' % g_err)
    print('best depth for gini : %d' % g_depth)

    print('\nENTROPY criterion:')
    print('Best achieved error : %.4f' % e_err)
    print('best depth for entropy is %d' % e_depth)
    
    return g_depth,e_depth;

g_depth,e_depth = gini_entropy_plot(X_train,y_train,X_test,y_test,maxdepth = 15)
```

```python
# Plotting the train Data
plt.figure(figsize = (6,6))
tp.plot_2d(X_train,y_train)
plt.title("Training data")
plt.xlabel("feature 1")
plt.ylabel("feature 2")
plt.show()
```

```python
# Plotting the ground truth about Test data 
plt.figure(figsize = (8,4))
plt.subplot(121)
tp.plot_2d(X_test,y_test)
plt.title("Test data : Ground Truth")
plt.xlabel("feature 1")
plt.ylabel("feature 2")
plt.subplot(122)
tp.plot_2d(X_test,y_test)
plt.title("Test data : Ground Truth")
plt.xlabel("feature 1")
plt.ylabel("feature 2")
plt.show()

# We refit our classifiers with fixed depth : We keep the seemingly best depth for each case
plt.figure(figsize = (8,4))
#GINI
g_clf = tree.DecisionTreeClassifier(criterion="gini",max_depth=g_depth,random_state=25)
g_clf = g_clf.fit(X_train,y_train)
y_pred = g_clf.predict(X_test)
g_err = 1 - accuracy_score(y_test,y_pred)
plt.subplot(121)
tp.plot_2d(X_test,y_pred)
plt.title("Test data classification with gini_clf")
plt.xlabel("feature 1")
plt.ylabel("feature 2")
#Entropy
e_clf = tree.DecisionTreeClassifier(criterion="entropy",max_depth=e_depth,random_state=25)
e_clf = e_clf.fit(X_train,y_train)
y_pred = e_clf.predict(X_test)
e_err = 1 - accuracy_score(y_test,y_pred)
plt.subplot(122)
tp.plot_2d(X_test,y_pred)
plt.title("Test data classification with entropy_clf")
plt.xlabel("feature 1")
plt.ylabel("feature 2")
# Plotting the predicted classifications
plt.show()
```

#### Displaying the classifier obtained using the depth that minimizes the error percentage with the entropy

The error for this test sample is higher then the error for the previous test set. This was expected since the fixed hyperparameter "max_depth" was chosen among the tested values such that it minimizes the error percentage on the test set. Thus, the built decision tree is biased over the training data for prediction and classification and the choice of the max_depth parameter is biased over the used test set.

We might consider a cross validation process for the selection of the hyper parameter instead of basing our choice on the error rate on the first test set.
